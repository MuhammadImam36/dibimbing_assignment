```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaProducer
from faker import Faker
from time import sleep
from datetime import datetime, timedelta
```

```python
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)

kafka_host = os.getenv("KAFKA_HOST")
kafka_topic = os.getenv("KAFKA_TOPIC_NAME")
```

```python
producer = KafkaProducer(bootstrap_servers=f"{kafka_host}:9092")
faker = Faker()
```

```python
def get_data():
    now = datetime.now()
    unix_time = faker.unix_time(start_datetime = now - timedelta(minutes = 60), end_datetime = now + timedelta(minutes = 180))
    
    return [
        str(uuid.uuid4()),
        faker.random_int(min=1, max=100),
        faker.random_element(elements=("Chair", "Table", "Desk", "Sofa", "Bed")),
        faker.safe_color_name(),
        faker.random_int(min=100, max=150000),
        datetime.fromtimestamp(unix_time).strftime('%H:%M:%S')
    ]
```

```python
while True:
    columns = [
        "order_id",
        "customer_id",
        "furniture",
        "color",
        "price",
        "ts",
    ]
    data_list = get_data()
    json_data = dict(zip(columns, data_list))
    _payload = json.dumps(json_data).encode("utf-8")
    print(_payload, flush=True)
    response = producer.send(topic=kafka_topic, value=_payload)
    print(response.get())
    sleep(3)
```
